## Solutions

```python
import pandas as pd
import pyro
import torch
import torch.nn
import matplotlib.pyplot as plt
from IPython.display import HTML
import pyro.optim as optim
import pyro.distributions as dist
from torch.distributions import constraints, Normal
from pyro.infer import SVI, Trace_ELBO
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from tqdm import tqdm_notebook as tqdm

pyro.enable_validation(True)

num_iterations = 1000
softplus = torch.nn.Softplus()

dataset = pd.read_csv('wine.csv', sep=",", header=None)
cls_count = 3
attr_count = 13


def score(actual, predicted):
    return {
        # average = macro means averaging score for each label without label weight
        'Accuracy': accuracy_score(actual, predicted),  # number of correctly classified samples divided by all samples
        'Precision': precision_score(actual, predicted, average='macro'),
        # tp / (tp + fp) where tp is the number of true positives and fp the number of false positives.
        # The precision is intuitively the ability of the classifier not to label as positive a sample that is negative
        'Recall': recall_score(actual, predicted, average='macro'),
        # tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives.
        # The recall is intuitively the ability of the classifier to find all the positive samples.
        'F1_Score': f1_score(actual, predicted, average='macro')
        # F1 = 2 * (precision * recall) / (precision + recall)
    }


# P(X|Y)
def get_probs(data, mean, std, pk):
    def sample_attr_prob(attr, clz):
        return torch.exp(Normal(loc=mean[attr, clz], scale=softplus(std[attr, clz])).log_prob(data[:, attr]))

    p_x_c = torch.stack(
        [
            torch.prod(torch.stack(
                [sample_attr_prob(attr_ind, clz_ind) for attr_ind in range(attr_count)]
            ), dim=0) for clz_ind in range(cls_count)
        ])
    probs = p_x_c.transpose(0, 1) * pk.squeeze()
    return probs


def model(x_data, y_data):
    p_cls_prior = dist.Dirichlet(torch.ones(cls_count, ))
    p_cls_sample = pyro.sample('p_cls', p_cls_prior)
    with pyro.plate('cls', size=cls_count):
        with pyro.plate('attr', size=attr_count):
            mean_prior = dist.Normal(loc=torch.ones(attr_count, cls_count) * 2,
                                     scale=torch.ones(attr_count, cls_count) * 2)
            std_prior = dist.Normal(loc=torch.ones(attr_count, cls_count) * 0.5,
                                    scale=torch.ones(attr_count, cls_count))
            mean_sample = pyro.sample('attr_mean', mean_prior)
            std_sample = pyro.sample('attr_std', std_prior)

    # https://scikit-learn.org/stable/modules/naive_bayes.html
    yhat = get_probs(x_data, mean_sample, std_sample, p_cls_sample)
    with pyro.plate('plate', len(y_data)):
        pyro.sample("obs", dist.Categorical(probs=yhat), obs=y_data)


def guide(x_data, y_data):
    mean_loc = pyro.param("mean_loc", torch.ones(attr_count, cls_count) * .5)
    mean_scale = pyro.param("mean_scale", torch.ones(attr_count, cls_count), constraint=constraints.positive)
    std_loc = pyro.param("std_loc", torch.ones(attr_count, cls_count) * .6)
    std_scale = pyro.param("std_scale", torch.ones(attr_count, cls_count) * .5, constraint=constraints.positive)
    # dirichlet distr concentration
    p_cls_cnctr = pyro.param("p_cls_cnctr", torch.ones(cls_count, ), constraint=constraints.interval(0.01, 10))

    pk_prior = dist.Dirichlet(p_cls_cnctr)
    pk_sample = pyro.sample('p_cls', pk_prior)
    with pyro.plate('cls', size=cls_count):
        with pyro.plate('attr', size=attr_count):
            mean = dist.Normal(loc=mean_loc, scale=mean_scale)
            std = dist.Normal(loc=std_loc, scale=std_scale)
            mean_sample = pyro.sample('attr_mean', mean)
            std_sample = pyro.sample('attr_std', std)

    return mean_sample, std_sample, pk_sample


min_max_scaler = preprocessing.MinMaxScaler()
attrs = torch.Tensor(min_max_scaler.fit_transform(dataset.iloc[:, 0:-1].values))
# attrs = torch.Tensor(dataset.iloc[:, 0:-1].values)
cls = torch.Tensor(dataset.iloc[:, -1].values)

pyro.clear_param_store()
lr = 0.05
optimizer = optim.Adam({"lr": lr})
svi = SVI(model, guide, optimizer, loss=Trace_ELBO())

tq = tqdm(range(num_iterations))
losses = []
for j in tq:
    loss = svi.step(attrs, cls)
    losses.append(loss)
    tq.set_postfix_str(f"Loss: {loss}")
plt.plot(losses)
plt.ylabel("ELBO loss")
plt.xlabel("iteration")
plt.show()

params = guide(None, None)
print(f"Means: \n {params[0]}")
print(f"Std: \n{softplus(params[1])}")
print(f"Class prob: \n{params[2]}")

prob_y = get_probs(attrs, params[0], params[1], params[2])
pred_y = torch.argmax(prob_y, dim=1)
print(f"Actual: \n{cls}")
print(f"Predicted: \n{pred_y}")
scr = score(cls, pred_y)
print(f"Score: \n {scr}")
```
